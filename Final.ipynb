# Finding Hidden Gems

Using a method called k-Means Clustering, I have analyzed every player's stats and grouped them in tiers based on how similar they are. By comparing these tiers to what players are actually being paid, we can find hidden gem players who perform like stars but are paid like bench warmers.

### **How the Model Works:**

I used a k-Means Clustering algorithm, which looks at every player simultaneously across several key metrics:

- **Production:** Total Points (PTS), Rebounds (TRB), and Assists (AST) to measure raw output
- **Efficiency:** Effective Field Goal Percentage (eFG%), Three-Point Percentage (3P%), and Free Throw Percentage (FT%) to ensure we're picking players who consistently shoot well
- **Defense:** Steals (STL) and Blocks (BLK) to find the players who disrupt the opposition
- **Opportunity:** Total Minutes Played (MP) to see how much a coach trusts them to play

I scaled these variables so that a high free-throw percentage (which is a decimal) wouldn't be ignored next to a high point total (which is in the thousands). By grouping players into just two main clusters, the model separated the league into "High Performance" (Cluster 0) and "Lower Performance" (Cluster 1). Our goal is to find the anomalies: the players in Cluster 0 who are being paid Cluster 1 salaries.

### **Player Recommendations**
#### **The Good Choices:** These players are in the High Performance Cluster but are being paid like bench players

**Russel Westbrook:** His performance metric is very high, yet his salary remains extremely low relative to his impact.

**Keyonte George:** Similar to Westbrook, he is clustered with the stars of the league while he is being paid relatively lower than others of similar performance. 

**Colin Gillespie:** He shows up as a high performer in both the cluster and through performance score. He is a "star" in the model's eyes, but his contract is significantly lower than the other stars he is clustered with.

#### **The Bad Choices:** These players are "traps." They are in the High Performance Cluster because they play a lot of minutes and score points, but their salary is relatively higher than their performance. 

**Luguentz Dort:** There are several players within the Lower Performance Cluster who have better performance, but lower salary than him. The price for his performance is not worth it. 

**Zach LaVine:** While his performance score is quite high, he is one of the highest paid players among the league despite having not so unqiue skill. 

**Jalen Suggs:** Similar to both Dort and LaVine, his performance is quite good, but is being paid too much for his skill.

#### **The Maybe Choices:** They are currently in the Lower Performance Cluster, but the Efficiency vs. Opportunity chart shows they may have potential. 

**Jaxson Hayes:** He has one of the highest Effective Field Goal percentages and has a good amount of points scored. He also is being paid on the lower end of the league. 

**Jericho Sims:** He is one of the most efficient shooters in the league and has a lower salary. While his points are low, his skill could be leveraged in a team. 

**Walker Kessler**: He has a high eFG%, but doesn't have a high volume of points nor minutes played. Given the chance to be on the court more, he could rise. 

### **Conclusion**
By avoiding the "Salary Traps" like Dort and LaVine, and instead targeting high-efficiency, low-cost talent like Westbrook or Keyonte, we can build a resilient roster wih cheaper players that can compete for the playoffs.
